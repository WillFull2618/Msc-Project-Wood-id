```python
from matplotlib import pyplot as plt
import numpy as np
from PIL import Image
%matplotlib inline
```

```python
#loading training and testing label files
Train_labels = np.load('Wood_Data/Train_labels.npy')
Test_labels  = np.load('Wood_Data/Test_labels.npy')
```

```python
#training images
#returns image as numpy array
def loadtrain(img_num):
    path = 'Wood_Data/Train/{}.jpg'.format(str(img_num))
    print('Accessing: {}'.format(path))
    img = Image.open(path)
    arr = np.array(img)
    return arr
```

```python
#testing images
#returns image as numpy array
def loadtest(img_num):
    path = 'Wood_Data/Test/{}.jpg'.format(str(img_num))
    print('Accessing: {}'.format(path))
    img = Image.open(path)
    arr = np.array(img)
    return arr
```

```python
#example of loading a Training image
img_number = 3564 #max 5708
img = loadtrain(img_number)
plt.imshow(img)
plt.show()
print('Label: {}'.format(str(Train_labels[img_number-1])))
```

```python
#example of loading a Test image
img_number = 2500# max 2836
img = loadtest(img_number)
plt.imshow(img)
plt.show()
print('Label: {}'.format(str(Test_labels[img_number-1])))
```
